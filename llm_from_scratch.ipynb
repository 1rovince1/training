```python
import torch.nn.functional as F

y = torch.tensor([1.])
x1 = torch.tensor([1.1])
w1 = torch.tensor([0.84])
b = torch.tensor([2.3])

z = x1 * w1 + b
a = torch.sigmoid(z)

loss = F.binary_cross_entropy(a, y)
```

```python
loss
```

```python
import torch.nn.functional as F
from torch.autograd import grad

y = torch.tensor([1.])
x1 = torch.tensor([1.1])
w1 = torch.tensor([0.84], requires_grad=True)
b = torch.tensor([2.3], requires_grad=True)

z = x1 * w1 + b
a = torch.sigmoid(z)

loss = F.binary_cross_entropy(a, y)

grad_L_w1 = grad(loss, w1, retain_graph=True)
grad_L_b = grad(loss, b, retain_graph=True)
```

```python
print(grad_L_w1, grad_L_b)
```

```python
loss.backward()
```

```python
print(w1.grad, b.grad)
```

```python
class NeuralNetwork(torch.nn.Module):
    def __init__(self, input_units, output_units):
        super().__init__()

        self.layers = torch.nn.Sequential(
            torch.nn.Linear(input_units, 30),
            torch.nn.ReLU(),
            torch.nn.Linear(30, 20),
            torch.nn.ReLU(),
            torch.nn.Linear(20, output_units)
        )

    def forward(self, x):
        logits = self.layers(x)
        return logits
```

```python
model = NeuralNetwork(50, 3)
```

```python
print(model)
```

```python
print(sum([p.numel() for p in model.parameters() if p.requires_grad]))
print(len([p for p in model.parameters()]))
print([p for p in model.parameters()])
```

```python
print(model.layers[0].weight)
print(model.layers[0].weight.shape)
print(model.layers[0].bias)
```

```python
torch.manual_seed(123)
model = NeuralNetwork(50, 3)
print(model.layers[0].weight)
```

```python
torch.manual_seed(123)
X = torch.rand((1, 50))
out = model(X)
print(out)
```
